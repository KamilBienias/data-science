# **Python Pandas Tutorial – modyfikacje kolumn**

```
import pandas as pd
# header=2 bo nazwy kolumn są w 3 wierszu
kostium = pd.read_csv('Halloween.csv', header=2)
kostium.head()
```

## **Dodanie nowej kolumny**

Utworzenie kolejnej kolumny w Pandas jest banalne. Wystarczy w nawiasach [ ] podać nazwę nowej i zostanie ona utworzona automatycznie.

```
kostium['Nowa'] = '0'
kostium
```

## **Edycja wartości kolumn**

Poznaliśmy funkcję loc, która umożliwia nam edycję wartości, oraz maski które również do tego celu możemy wykorzystać. Jeżeli chcielibyśmy zmienić wartość kolumny 'Nowa' na 1, tam gdzie najpopularniejszym kostiumem jest 'Harley Quinn', wystarczy że zrobimy to tak:

```
kostium.loc[kostium['1'] == 'Harley Quinn', ["Nowa"]] = "1"
kostium
```

## **Uporządkowanie kolumn**

### **Zmiana nazwy kolumn**

Nazwy naszych kolumn nie są jeszcze zbyt przyjazne. Możemy je łatwo zmienić za pomocą funkcji rename, która przyjmuje parametr columns, podając w postaci słownika nazwy starych oraz nowych kolumn.

```
# ta zmiana nazw kolumn nie jest na stałe, bo nie była przypisana do DataFrame
kostium.rename(columns={'1':'Pierwszy', '2':'Drugi'})
```

### **Kasowanie kolumny**

Skasować kolumnę możemy za pomocą funkcji drop. Funkcja ta umożliwia nam kasowanie zarówno rekordów jak i kolumn. Aby skasować rekord wystarczy podać jako parametr pozycję w indeksie. Jeżeli chcemy skasować kolumnę, podajemy nazwę kolumny, oraz informacje że chcemy skasować kolumnę czyli axis = 1

```
kostium.drop('Nowa', axis=1)
```

```
# Usuwam wiersz o indeksie 2. Kolumna "Nowa" jest nadal, bo jej usunięcie nie było przypisane do df.
# Podobnie wiersz 2 nie został na stałe usunięty.
kostium.drop(2)
```

## **Scalanie i rozdzielanie kolumn**

### **Scalanie**

Aby połączyć dwie lub więcej kolumn w jedną, wystarczy utworzyć nową kolumnę na bazie już istniejących. Przykładowo w ten sposób:

```
kostium['Połączone'] = kostium['3'] + ' | ' + kostium['4']
kostium.head()
```

### **Rozdzielanie**

Mamy nową kolumnę 'Połączone'. Dość częstą sytuacją jest rozdzielanie tego typu kolumn, na kolumny osobne. Często na podstawie konkretnego znaku w ciągu znaków. Możemy to zrobić na kilka sposobów. O kolejnych będziemy się dowiadywać w następnych lekcjach, jednak popularnym sposobem jest skorzystanie z funkcji split, znanej nam z operacji na łańcuchach znaków:

```
# Jeśli nie podam seperatora, to domyślnie rozdziela po spacji.
# Bez expand=True na wyjściu dostalibyśmy Series
kostium[['Trzeci', 'Czwarty']] = kostium.Połączone.str.split('|', expand=True)
kostium.head()
```

## **Ćwiczenia**

3. Utworzyć nowy zbiór danych, który nie zawiera regionów w których najpopularniejszym strojem był 'Angel'

```
# 1. Kasuję na stałe kolumny 3 i 4
kostium = kostium.drop(["3", "4"], axis=1)
kostium.head()
```

```
# 2. Zmieniam na stałe kolumnę "5" na "Piąty"
kostium = kostium.rename(columns={"5": "Piąty"})
kostium.head()
```

```
# 3. Najpierw z najpopularniejszym Angel
kostium_z_najpopularnieszym_Angel = kostium.loc[kostium["1"] == "Angel"]
kostium_z_najpopularnieszym_Angel
```

```
# 3. Tereaz wyrzucam te dwa wiersze z DataFrame kostium
wyrzucone_indeksy = kostium_z_najpopularnieszym_Angel.index
kostium_bez_najpopularnieszego_Angel = kostium.drop(wyrzucone_indeksy)
# już nie ma indeksów 25 i 47
kostium_bez_najpopularnieszego_Angel
```
