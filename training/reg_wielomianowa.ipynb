```
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

# ilość próbek
m = 100

# podobny do przykładu z dokumentacji,
# ale uogólniłem na m próbek i dorzuciłem y
X = np.arange(m*2).reshape(m, 2)
X
# X1, X2
```

```
# tak było na początku
# X1 = np.array([X[0][0], X[1][0], X[2][0]])

X1_list = []
for i in range(len(X)):
    element = X[i][0]
    X1_list.append(element)
X1 = np.array(X1_list)
X1
```

```
# tak było na początku
# X1 = np.array([X[0][1], X[1][1], X[2][1])

X2_list = []
for i in range(len(X)):
    element = X[i][1]
    X2_list.append(element)
X2 = np.array(X2_list)
X2
```

```
# y bez szumu
y_clean = -X1*X1 - X1*X2 + X2*X2
y_clean
```

```
# szumem jest m liczb z rozkładu N(0,200)
noise = 200 * np.random.randn(m)
noise
```

```
# y_true
y_true = -X1*X1 - X1*X2 + X2*X2 + noise
y_true
```

```
poly = PolynomialFeatures(2)
X_poly_2 = poly.fit_transform(X)
X_poly_2
# 1, X1, X2, X1**2, X1*X2, X2**2
```

```
from sklearn.linear_model import LinearRegression

regressor_poly_2 = LinearRegression()
regressor_poly_2.fit(X_poly_2, y_true)

y_pred_2 = regressor_poly_2.predict(X_poly_2)
y_pred_2
```

```
import pandas as pd

df = pd.DataFrame({
    "X1": X1,
    "X2": X2,
    "y_clean": y_clean,
    "y_true": y_true,
    "y_pred": y_pred_2,
    "y_pred_2 - y_clean": y_pred_2 - y_clean
})
df
```

```
import plotly.express as px

fig = px.scatter_3d(df, x='X1', y='X2', z='y_true', labels={"y_true": "prawdziwy"})
fig2 = px.line_3d(df, x='X1', y='X2', z='y_pred', labels={"y_pred": "predykcja"})
fig.update_traces(marker_color='yellow')
fig.add_trace(fig2.data[0])
fig.show()
# trzeba przybliżyć żeby zobaczyć niebieską linię predykcji
```
