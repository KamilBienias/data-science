### Porównanie do benchmarku (czyli jakby średniej całego rynku)

```python
# jeszcze raz Apple
aapl = pdr.get_data_yahoo("AAPL", start=dt.datetime(2019, 1, 1))
# szeroki rynek w USA
sp500 = pdr.get_data_yahoo("^GSPC", start=dt.datetime(2019, 1, 1))
```

```python
aapl["pct_change"] = aapl["Adj Close"].pct_change()
sp500["pct_change"] = sp500["Adj Close"].pct_change()
```

```python
sp500
```

```python
# połączy dwie df ze sobą
# keys to indeks hierarchiczny. Wstawiam nazwy obu df1 i df2 (nie muszą być takie same jak aapl i sp500)
# names to nazwy indeksu hierarchicznego oraz indeksu zwykłego
portfolio = pd.concat([aapl, sp500], keys=['aapl', 'sp500'], names=['Ticker', 'Date'])
portfolio
```

```python
# sam usuwam Date z indeksu i jest jako kolumna
portfolio = portfolio.reset_index()
portfolio
```

```python
# na filmie było [["Adj Close"]] ale rzucało błąd że nie ma Date
# Sprawdzałem, że ta samo będzie z pivot_table
portfolio = portfolio[["Adj Close", "Date", "Ticker"]].reset_index().pivot(index="Date", columns="Ticker", values="Adj Close")
portfolio
```

```python
portfolio.info()
```

```python
# stopy zwrotu, sharex nic nie zmieniło
# Apple jest szerszy czyli ma większe odchylenie standardowe,
# czyli większe ryzyko w inwestowanie w tą spółkę.
# Ale to normalne jak się porównuje do całego rynku sp500
portfolio.pct_change().hist(bins=50, sharex=True)
```

```python
# skumulowana stopa inwestycji w Apple i cały rynek.
# (1 + portfolio.pct_change()) to stopy zwrotu powiększone o 1
# cumprod() wymnaża i sumuje przez siebie stopy zwrotu. 
# Pierwsza data 2019-01-02 ma pusete miejsca NaN, 
# dlatego wypełnia fillna(1)
# Widać, że Apple radzi sobie lepiej niż cały rynek
(1 + portfolio.pct_change()).cumprod().fillna(1).plot()
```

## **Tworzymy nasz portfel**

```python
# stocks to lista spółek, które chcę mieć w portfelu
def make_portfolio(stocks, start_date, end_date):

    # Download data
    def get_data(stock):
        return pdr.get_data_yahoo(stock, start=start_date, end=end_date)
    
    # działa funkcją get_data na każdy element ze zbioru stocks
    stock_data = map(get_data, stocks)

    full_dataframe = pd.concat(stock_data, keys=stocks, names=["Stock", "Date"])

    return full_dataframe[["Adj Close"]].reset_index().pivot(index="Date", columns="Stock", values="Adj Close")
```

```python
# ostatnia to nasdac czyli indeks spółek technologicznych (nasz benchmark)
stocks = ["TSLA", "FB", "NFLX", "GOOG", "^IXIC"]
portfolio = make_portfolio(stocks=stocks, start_date=dt.datetime(2019, 1, 1), end_date=dt.datetime(2020, 4, 30))
```

```python
portfolio
```

### Wizualizacja porównania z benchmarkiem

```python
def compare_to_benchmark(portfolio, benchmark):

    benchmark_color_dict = {benchmark: "red"}

    plt.figure(figsize=(15, 20))

    # reszta spółek ma losowy kolor, bo jak nie ma klucza w słowniku to wybiera kolor 
    return (1 + portfolio.pct_change()).cumprod().fillna(1).plot(color=[benchmark_color_dict.get(x, np.random.rand(3)) for x in portfolio.columns])
```

```python
compare_to_benchmark(portfolio=portfolio, benchmark="^IXIC")
```
