### <a name='0'></a> Import bibliotek

````
print("############################################################")
print("Etap9. Odcinek: PCA - Analiza Głównych Składowych - breast cancer data")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px

np.set_printoptions(precision=4, suppress=True, linewidth=150)
````

### <a name='1'></a> Załadowanie danych - breast cancer

````
from sklearn.datasets import load_breast_cancer

raw_data = load_breast_cancer()
all_data = raw_data.copy()
data = all_data['data']
target = all_data['target']
data[:3]
````

````
target[:30]
````

````
# chce zredukować liczbę zmiennych z 30
data.shape
````

### <a name='2'></a> Standaryzacja

````
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_std = scaler.fit_transform(data)
data_std[:3]
````

### <a name='3'></a> PCA - 2 komponenty

````
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_std)
data_pca[:5]
````

````
pca_2 = pd.DataFrame(data={'pca_1': data_pca[:, 0], 'pca_2': data_pca[:, 1], 'class': target})
print("Przed zamianą kolumny 'class'")
print(pca_2)
print()
print("Zamienia 0 na Beingn, a 1 na Malignant")
pca_2.replace(0, 'Benign', inplace=True)
pca_2.replace(1, 'Malignant', inplace=True)
print(pca_2)
````

````
print("Skumulowana wyjaśniona wariancja")
# jedna kolumna explained_variance_ratio przy pomocy atrybutu explained_variance_ratio_ obiektu pca
results = pd.DataFrame(data={'explained_variance_ratio': pca.explained_variance_ratio_})
# dodaje kolumnę cumulative, która jest skumulowanymi liczbami z kolumny explained_variance_ratio
results['cumulative'] = results['explained_variance_ratio'].cumsum()
# kolumna compontnt jest ma wartości o 1 większe od indexu
results['component'] = results.index + 1
results
````

````
fig = go.Figure(data=[go.Bar(x=results['component'], y=results['explained_variance_ratio'], name='explained_variance_ratio'),
                      go.Scatter(x=results['component'], y=results['cumulative'], name='cumulative')],
                layout=go.Layout(title='PCA - 2 components', width=900, template='plotly_dark'))
fig.show()
````

````
# sam robię w matplotlib bo na github nie widać plotly
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8.0,6.0))
ax = fig.add_subplot(111)
ax.bar(results['component'], results['explained_variance_ratio'],
label='explained variance ratio')
ax.plot(results['component'], results['cumulative'], marker='o', label='cumulative explained variance', c="red")
plt.legend(loc='center right')
plt.show()
````

````
# wykres punktowy w przestrzeni R2 naszych dwóch głównych składowych
px.scatter(pca_2, 'pca_1', 'pca_2', color=pca_2['class'], width=950, template='plotly_dark')
````

````
# sam robię w matplotlib bo na github nie widać plotly
# wykres punktowy w przestrzeni R2 naszych dwóch głównych składowych
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8.0,6.0))
ax = fig.add_subplot(111)
punktowy = ax.scatter(pca_2['pca_1'], pca_2['pca_2'], c=raw_data['target'])
plt.xlabel("pca_1")
plt.ylabel("pca_2")
# bez poniższego wyświetlałoby legendę do każdego z 150 punktów osobno. Wzięte z:
# https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/scatter_with_legend.html
# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*punktowy.legend_elements(),
                    loc="lower left", title="Classes")
ax.add_artist(legend1)
plt.show()

# 0 to Beingn (łagodny)
# 1 to Malignant (złośliwy)
````

### <a name='5'></a> Zbiór danych MNIST

````
print("############################################################")
print("Etap9. Odcinek: PCA - Analiza Głównych Składowych - MNIST data")

from keras.datasets import mnist

# próbkami są zdjęcia 28 na 28 pikseli
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
````

Obcięcie danych do pierwszych 5000 zdjęć

````
# wystarczy pierwsze 5000 próbek
X_train = X_train[:5000]
y_train = y_train[:5000]
````

````
# pierwsza próbka
X_train[0]
````

````
# pięć pierwszych wartości klas ze zbioru treningowego. Pierwsza liczba to 5
y_train[:5]
````

Wizualizacja danych

````
plt.figure(figsize=(12, 8))
for i in range(8):
    plt.subplot(240 + i + 1)
    plt.imshow(X_train[i], cmap='gray_r')
    plt.title(y_train[i], color='white', fontsize=17)
    plt.axis('off')
plt.show()
````

Standaryzacja 

````
X_train = X_train / 255.
X_test = X_test / 255.
X_train.shape
````

Wypłaszczenie obrazów

````
X_train = X_train.reshape(-1, 28 * 28)
X_train.shape
````

PCA - 3 komponenty

````
pca = PCA(n_components=3)

X_train_pca = pca.fit_transform(X_train)
X_train_pca[:5]
````

Wyjaśniona wariancja

````
results = pd.DataFrame(data={'explained_variance_ratio': pca.explained_variance_ratio_})
results['cumulative'] = results['explained_variance_ratio'].cumsum()
results['component'] = results.index + 1
results

# całkowita wyjaśniona wariancja przy pomocy trzech głównych składowych to tylko 23%
````

````
fig = go.Figure(data=[go.Bar(x=results['component'], y=results['explained_variance_ratio'], name='explained_variance_ratio'),
                      go.Scatter(x=results['component'], y=results['cumulative'], name='cumulative')],
                layout=go.Layout(title='PCA - 3 komponenty', width=950, template='plotly_dark'))
fig.show()
````

````
# sam robię w matplotlib bo na github nie widać plotly
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8.0,6.0))
ax = fig.add_subplot(111)
ax.bar(results['component'], results['explained_variance_ratio'],
label='explained variance ratio')
ax.plot(results['component'], results['cumulative'], marker='o', label='cumulative explained variance', c="red")
plt.legend(loc='center right')
plt.show()
````

````
X_train_pca_df = pd.DataFrame(np.c_[X_train_pca, y_train], columns=['pca_1', 'pca_2', 'pca_3', 'class'])
# zamienia 'class' na string
X_train_pca_df['class'] = X_train_pca_df['class'].astype('str')
X_train_pca_df.head()

# numer klasy to zarazem nazwa cechy
````

Wizualizacja 3D głównych komponentów

````
px.scatter_3d(X_train_pca_df, x='pca_1', y='pca_2', z='pca_3', color='class', 
              symbol='class', opacity=0.7, size_max=10, width=900, height=700, 
              template='plotly_dark', title='PCA - MNIST dataset')
````

````
# sam robię w matplotlib bo na github nie widać plotly
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16.0,12.0))
ax = fig.add_subplot(111, projection='3d')
punktowy_3d = ax.scatter(X_train_pca_df['pca_1'], X_train_pca_df['pca_2'], X_train_pca_df['pca_3'], c=y_train)
ax.set_xlabel('pca_1')
ax.set_ylabel('pca_2')
ax.set_zlabel('pca_3')
# bez poniższego wyświetlałoby legendę do każdego z 150 punktów osobno. Wzięte z:
# https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/scatter_with_legend.html
# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*punktowy_3d.legend_elements(),
                    loc="lower left", title="Classes")
ax.add_artist(legend1)
plt.show()
````

````
# chce dopasować pca aby wyjaśniało 95% wariancji (zamiast ilości komponentów podaje ułamek)
pca = PCA(n_components=0.95)

X_train_pca = pca.fit_transform(X_train)
X_train_pca[:1]
````

````
# tyle potrzeba komponentów spośrów wyjściowych 784. Czyli148 zmiennych trzyma 95% rozrzutu danych
pca.n_components_
````

Wyjaśniona wariancja

````
results = pd.DataFrame(data={'explained_variance_ratio': pca.explained_variance_ratio_})
results['cumulative'] = results['explained_variance_ratio'].cumsum()
results['component'] = results.index + 1
results
````

````
# wyjaśniona wariancja w zależności od ilości komponentów
fig = go.Figure(data=[go.Bar(x=results['component'], y=results['explained_variance_ratio'], name='explained_variance_ratio'),
                      go.Scatter(x=results['component'], y=results['cumulative'], name='cumulative')],
                layout=go.Layout(title='PCA - 3 komponenty', width=900, template='plotly_dark'))
fig.show()
````

````
# sam robię w matplotlib bo na github nie widać plotly
# Wyjaśniona wariancja w zależności od ilości komponentów
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16.0,12.0))
ax = fig.add_subplot(111)
ax.bar(results['component'], results['explained_variance_ratio'],
label='explained variance ratio')
ax.plot(results['component'], results['cumulative'], marker='o', label='cumulative explained variance', c="red")
plt.legend(loc='center right')
plt.show()
````
