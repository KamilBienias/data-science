```
df1 = pd.DataFrame(np.random.rand(10, 4), columns=list('abcd'))
df1
```

```
df2 = pd.DataFrame(np.random.rand(10, 4), columns=list('abcd'))
df2
```

```
df3 = pd.DataFrame(np.random.rand(10, 4), columns=list('abcd'))
df3
```

```
# %%
# sort_index() sortuje indeksy po złączeniu 0,0,1,1,2,2,... 
# reset_index() robi z indeksów kolumnę o nazwie 'index', a jako indeksy tworzy 0,1,2,3...
df = df1.append(df2).sort_index().reset_index()
df
```

```
# usuwa kolumnę 'index'
del df['index']
df
```

```
# %% robi to samo, ale zamiast del używa drop()
df = df1.append(df2).sort_index().reset_index().drop('index', axis=1)
df
```

```
# %% połączenie df1 i df2 ignorując początkowe indeksy
df = df1.append(df2, ignore_index=True)
df
```
