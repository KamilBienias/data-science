```
ten = ten_pl.copy()
plw = plw_pl.copy()
print("ten =")
print(ten.head())
print()
print("plw =")
print(plw.head())
```

```
# zamiana nazw kolumn z polskich na angielskie
ten.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
plw.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

# sam to wyświetlam
print("ten =")
print(ten.head())
print(ten.tail())
print()
print()
print("plw =")
print(plw.head())
print(plw["2018-05-11":"2018-05-17"])
print(plw.tail())
```

```
# %% merge on index, inner. Pokazuje tylko wiersze wspólne czyli od 2018-05-11
df = pd.merge(ten, plw, how='inner', left_index=True, right_index=True)
df.head()
```

```
# %% zmiana suffiksów z x na TEN i z y na PLW
df = pd.merge(ten, plw, how='inner', left_index=True, right_index=True,
              suffixes=('_TEN', '_PLW'))
df.head()
```

```
# %% merge on index, outer. Pokazuje wszystkie wiersze, czyli od 2016-10-19
df = pd.merge(ten, plw, how='outer', left_index=True, right_index=True)
df
```

```
# %% merge on index, outer oraz zmiana suffixów
df = pd.merge(ten, plw, how='outer', left_index=True, right_index=True,
              suffixes=('_TEN', '_PLW'))
df
```

```
# %% generate daily changes. Generuje dzienne zmiany
ten['Change'] = ten['Close'] / ten['Close'].shift() - 1
plw['Change'] = plw['Close'] / plw['Close'].shift() - 1
```

```
# %% generate flag 1 if change > 0, 0 if change < 0
# W df o nazwie 'ten' wstawia kolumnę 'Flag', gdzie wartość True jeśli ten['Change'] > 0
# Tak samo dla df o nazwie 'plw'
ten['Flag'] = ten['Change'] > 0
plw['Flag'] = plw['Change'] > 0
```

```
# Zamienia True na 1, a False na 0
ten['Flag'] = ten['Flag'].apply(int)
plw['Flag'] = plw['Flag'].apply(int)
```

```
# %% merge table. Już są nowe kolumny Flag_TEN i Flag_PLW
df = pd.merge(ten, plw, how='inner', left_index=True, right_index=True,
              suffixes=('_TEN', '_PLW'))
df
```

```
# macierz korelacji
correlation_matrix = df.corr()
correlation_matrix
```

```
# %% select where Flag_TEN = Flag_PLW = 1
df_positive = df[(df['Flag_TEN'] == 1) & (df['Flag_PLW'] == 1)]
df_positive
```

```
# inny sposób na wyświetlenie wierszy, których obie flagi mają wartość 1
df_positive_2 = df.query('(Flag_TEN == 1) & (Flag_PLW == 1)')
df_positive_2
```
